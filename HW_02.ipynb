```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing 
from sklearn import metrics
```

```python
data = pd.read_csv("Treasury Squeeze raw score data.csv")
```

```python
data = pd.read_csv("Treasury Squeeze raw score data.csv").drop(columns=["rowindex"])
```

```python
data
```

```python
data.cov()
```

Squeeze will be the target values and rest features

```python
#Taking all columns as features
X = data.drop(columns=["squeeze","contract"])
#taking two columns as features namely price_crossing and price_distortion
X_2 = data.drop(columns=["contract","squeeze","roll_start","roll_heart","near_minus_next","ctd_last_first","ctd1_percent","delivery_cost","delivery_ratio"])
y = data["squeeze"]
y=y.astype('int')
y
```

Since column 1 is categorical data, we can drop it.

```python
# X = data.drop(columns=["contract","squeeze"])
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y, test_size=0.25, random_state=0)



```

```python
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions
for i in range(1,15):
    tree = DecisionTreeClassifier(criterion = 'gini', max_depth =i,random_state=1)
    tree.fit(X_train, y_train)
    y_pred = tree.predict(X_test)
    accuracy_1 = metrics.accuracy_score(y_test, y_pred)
#     print("Acccuracy of max_depth: ",i," is ", accuracy)
#     print("Training model using 2 features")
    tree_2 = DecisionTreeClassifier(criterion = 'gini', max_depth =i,random_state=1)
    tree_2.fit(X_train_2, y_train_2)
    y_pred = tree_2.predict(X_test_2)
    accuracy_2 = metrics.accuracy_score(y_test, y_pred)
    print("Acccuracy of max_depth: ",i," is ", accuracy_1," two features model ", accuracy_2)
# X_combined = np.vstack((X_train_2, X_test_2))
# y_combined = np.hstack((y_train_2, y_test_2))
# plot_decision_regions(X_combined, y_combined, clf = tree)
# plt.show()

```

Max depth of decision tree should be 4 as 5 onwards it decreses again. for two features model max_depth is 8.

```python
from sklearn.ensemble import RandomForestClassifier
for i in range(5,70,5):
    forest = RandomForestClassifier(criterion ='gini', n_estimators=i,random_state=1, n_jobs=2)
    forest.fit(X_train, y_train)
    y_pred = forest.predict(X_test)
    accuracy = metrics.accuracy_score(y_test, y_pred)
    print("Acccuracy of number_of_estimators: ",i," is ", accuracy)
# plot_decision_regions(X_combined, y_combined, clf=forest)
# plt.xlabel('price_crossing')
# plt.ylabel('price_distortion')
# plt.legend(loc='upper left')
# plt.show()



```

We can observe maximum accuracy for n_estimators = 40

```python
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i,p=2, metric = 'minkowski')
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = metrics.accuracy_score(y_test, y_pred)
    print("Acccuracy of number_of_neighbors: ",i," is ", accuracy)
# plot_decision_regions(X_combined, y_combined, clf=knn)
# plt.xlabel('price_crossing')
# plt.ylabel('price_distortion')
# plt.legend(loc='upper left')
# plt.show()
```

We observe maximum accuracy for number of neighbors = 13.
